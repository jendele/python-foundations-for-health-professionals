# Part A — Comprehensions

*Six exercises. Build new collections from existing ones, in one expression.*


### A1 — All wards (with duplicates)

*🟢 BASIC*

Build a list of all wards across all patients. Don't deduplicate — we want to see how often each ward appears.

```python
all_wards = [p["ward"] for p in patients]
print(all_wards)
print("Count:", len(all_wards))
```

**Why this works:** Standard list comprehension. Each patient contributes one ward to the new list.

### A2 — Unique wards

*🟢 BASIC*

Now make a SET of all unique wards. Same data, different shape.

```python
unique_wards = {p["ward"] for p in patients}
print(unique_wards)
print("Count:", len(unique_wards))
```

**Why this works:** Set comprehension. Duplicates disappear automatically.

### A3 — IDs of patients with chronic conditions

*🟡 INTERMEDIATE*

Build a list of patient IDs for patients who have AT LEAST ONE chronic condition.

```python
ids_with_conditions = [p["id"] for p in patients if p["chronic"]]
print(ids_with_conditions)
```

**Why this works:** The filter `if p['chronic']` exploits truthiness — empty lists are False, non-empty are True.

### A4 — Patient ID to age dict

*🟡 INTERMEDIATE*

Build a dictionary mapping patient ID to age.

```python
age_lookup = {p["id"]: p["age"] for p in patients}
print(age_lookup)
print("Age of P0006:", age_lookup["P0006"])
```

**Why this works:** Dict comprehension. Keys are patient IDs, values are ages.

### A5 — All distinct chronic conditions across all patients

*🟡 INTERMEDIATE*

Build a SET of all chronic condition names that appear anywhere in the patient list. Each patient has a list of conditions; we want every unique condition that appears in any patient.

```python
all_conditions = {cond for p in patients for cond in p["chronic"]}
print(all_conditions)
```

**Why this works:** Two for-clauses inside the comprehension: the outer iterates patients, the inner iterates that patient's conditions.

### A6 — Encounter cost per ward

*🔴 FAST-TRACK*

Build a dictionary `{ward: total_cost}` aggregating costs from `encounters`. Hint: this can't be done with a single comprehension easily; use a regular for-loop or set up an intermediate step.

```python
# Approach 1 — regular loop
total_cost_per_ward = {}
for e in encounters:
    total_cost_per_ward[e["ward"]] = total_cost_per_ward.get(e["ward"], 0) + e["cost_eur"]
print("Approach 1:", total_cost_per_ward)

# Approach 2 — dict comprehension over unique wards (more elegant but more code)
unique_wards = {e["ward"] for e in encounters}
total_cost_per_ward_v2 = {
    w: sum(e["cost_eur"] for e in encounters if e["ward"] == w)
    for w in unique_wards
}
print("Approach 2:", total_cost_per_ward_v2)
```

**Why this works:** Two ways to do this. Approach 1 (regular loop) is most idiomatic Python. Approach 2 (nested) shows the elegance of comprehensions but is slightly less efficient.

### B1 — average_age()

*🟢 BASIC*

Write a function `average_age(patient_list)` that returns the mean age across the given list.

```python
def average_age(patient_list):
    """Return mean age. Returns 0 if the list is empty."""
    if not patient_list:
        return 0
    ages = [p["age"] for p in patient_list]
    return sum(ages) / len(ages)

print("All:", average_age(patients))
print("Empty:", average_age([]))
```

**Why this works:** Simple aggregation function with a guard clause for empty input.

### B2 — count_in_ward()

*🟢 BASIC*

Write a function `count_in_ward(patient_list, ward_name)` that returns how many patients are in the given ward.

```python
def count_in_ward(patient_list, ward_name):
    """Count patients currently assigned to the given ward."""
    return sum(1 for p in patient_list if p["ward"] == ward_name)

print("Cardiology:", count_in_ward(patients, "Cardiology"))
print("ICU:", count_in_ward(patients, "ICU"))
print("Nephrology:", count_in_ward(patients, "Nephrology"))
```

**Why this works:** `sum(1 for ... if ...)` is a Pythonic counter. Equivalent to `len([... for ... if ...])` but uses less memory.

### B3 — high_risk() with default arguments

*🟡 INTERMEDIATE*

Write a function `high_risk(patient_list, min_age=65, min_conditions=2)` that returns the list of patients who are at least `min_age` years old AND have at least `min_conditions` chronic conditions.

```python
def high_risk(patient_list, min_age=65, min_conditions=2):
    """Patients aged >= min_age AND having >= min_conditions chronic conditions."""
    return [
        p for p in patient_list
        if p["age"] >= min_age and len(p["chronic"]) >= min_conditions
    ]

print("Default:", [p["id"] for p in high_risk(patients)])
print("More strict:", [p["id"] for p in high_risk(patients, min_age=80)])
print("Less strict:", [p["id"] for p in high_risk(patients, min_age=50, min_conditions=1)])
```

**Why this works:** Default arguments make the function flexible without forcing every caller to specify all parameters.

### B4 — describe_patient() — multi-line output, multi-return

*🟡 INTERMEDIATE*

Write a function `describe_patient(p)` that returns a TUPLE of two strings: `(headline, details)`. Headline like `'P0001 (67y, F)'`. Details like `'Cardiology · 2 chronic conditions: hypertension, diabetes_t2'`.

```python
def describe_patient(p):
    """Return (headline, details) for a patient."""
    headline = f"{p['id']} ({p['age']}y, {p['gender']})"
    if p["chronic"]:
        n = len(p["chronic"])
        cond_str = ", ".join(p["chronic"])
        details = f"{p['ward']} · {n} chronic condition{'s' if n > 1 else ''}: {cond_str}"
    else:
        details = f"{p['ward']} · no chronic conditions"
    return headline, details

for p in patients[:4]:
    headline, details = describe_patient(p)
    print(headline)
    print(f"  {details}")
    print()
```

**Why this works:** Multi-return via tuple. Note the use of `'s' if n > 1 else ''` to handle singular vs. plural.

### B5 — group_by_ward()

*🟡 INTERMEDIATE*

Write a function `group_by_ward(patient_list)` that returns a dictionary `{ward_name: [list_of_patients]}`. Each ward maps to the list of patients on that ward.

```python
def group_by_ward(patient_list):
    """Group patients into {ward: [patients]}."""
    groups = {}
    for p in patient_list:
        groups.setdefault(p["ward"], []).append(p)
    return groups

groups = group_by_ward(patients)
for ward, ps in groups.items():
    print(f"{ward}: {len(ps)} patients")
```

**Why this works:** `setdefault(key, [])` returns the existing list if the key is there, otherwise creates a new empty list and returns it. Then `.append()` works directly.

### B6 — total_cost_for_patient() with optional encounter list

*🔴 FAST-TRACK*

Write a function `total_cost_for_patient(patient_id, encounters_list=None)`. If `encounters_list` is `None`, use the global `encounters` variable. Returns the sum of `cost_eur` for all encounters of that patient.

```python
def total_cost_for_patient(patient_id, encounters_list=None):
    """Sum cost_eur for all encounters of patient_id.

    If encounters_list is None, use the global 'encounters'.
    """
    if encounters_list is None:
        encounters_list = encounters   # fall back to global

    relevant = [e for e in encounters_list if e["patient_id"] == patient_id]
    return sum(e["cost_eur"] for e in relevant)

print("P0001 (using global):", total_cost_for_patient("P0001"))
print("P0003 (using global):", total_cost_for_patient("P0003"))

subset = encounters[:5]
print("P0001 (using subset):", total_cost_for_patient("P0001", subset))
```

**Why this works:** The `None` default is the safe pattern. Inside the function, we explicitly use the global only if no argument was provided. Reading globals from inside a function is OK if it's intentional and documented.

### C3 — The mutable-default trap

*🟡 INTERMEDIATE*

Run this code. Why does the second call's output look wrong? Fix the function so each call starts with a fresh log.

```python
def add_log(message, log=None):
    if log is None:
        log = []
    log.append(message)
    return log

print(add_log("admit P0001"))   # ['admit P0001']
print(add_log("admit P0002"))   # ['admit P0002']
print(add_log("admit P0003"))   # ['admit P0003']

# To accumulate explicitly, pass the list yourself:
shared = []
shared = add_log("admit P0001", shared)
shared = add_log("admit P0002", shared)
shared = add_log("admit P0003", shared)
print("Accumulated:", shared)
```

**Why this works:** Default arguments evaluate ONCE at definition. With `log=[]`, all calls share that one list. With `log=None`, we create a fresh list inside the function.

*🟡 INTERMEDIATE*

Run the code. The output of `mean_age` looks wrong. Find the bug.

```python
def mean_age(patient_list):
    if not patient_list:
        return 0
    ages = [p["age"] for p in patient_list]
    return sum(ages) / len(ages)   # ← needed `return`

mean = mean_age(patients)
print("Mean age:", mean)
```

### C5 — Refactor: spaghetti to functions

*🔴 FAST-TRACK*

Below is a working but messy script. Refactor it into THREE functions: `parse_chronic_field(p)`, `is_high_risk(p)`, `summarise(patient_list)`. Test that the output is identical.

```python
def parse_chronic_field(p):
    """Add a 'chronic_list' field to p by splitting the 'chronic' string."""
    raw = p.get("chronic", "")
    p["chronic_list"] = raw.split(";") if raw else []
    return p

def is_high_risk(p, min_age=65, min_conditions=2):
    """True if p is old AND has multiple chronic conditions."""
    return p["age"] >= min_age and len(p["chronic_list"]) >= min_conditions

def summarise(patient_list):
    """Print totals."""
    high_risk = [p for p in patient_list if is_high_risk(p)]
    print(f"Total: {len(patient_list)}")
    print(f"High-risk: {len(high_risk)}")

# Use them
demo = [
    {"id": "X1", "age": 78, "chronic": "diabetes_t2;hypertension"},
    {"id": "X2", "age": 45, "chronic": ""},
    {"id": "X3", "age": 81, "chronic": "copd;heart_failure;ckd"},
    {"id": "X4", "age": 33, "chronic": "asthma"},
]
demo = [parse_chronic_field(p) for p in demo]
summarise(demo)
```

**Why this works:** Each function does one thing and has a clear name. is_high_risk is parameterised so we can reuse it with different thresholds without touching the function body.

### C6 — Build a tiny report

*🔴 FAST-TRACK*

Combine everything: write a function `report(patient_list, encounters_list)` that prints a complete hospital report with: total patients, total encounters, average age, high-risk count, total revenue, top 3 highest-cost encounters. Use the functions you wrote in earlier exercises.

```python
def average_age(patient_list):
    if not patient_list:
        return 0
    return sum(p["age"] for p in patient_list) / len(patient_list)

def high_risk(patient_list, min_age=65, min_conditions=2):
    return [p for p in patient_list
            if p["age"] >= min_age and len(p["chronic"]) >= min_conditions]

def report(patient_list, encounters_list):
    """Print a small hospital operations report."""
    print("=" * 50)
    print("Saint Mary's Hospital — Operations Snapshot")
    print("=" * 50)
    print(f"Total patients:    {len(patient_list)}")
    print(f"Total encounters:  {len(encounters_list)}")
    print(f"Average age:       {average_age(patient_list):.1f}")
    print(f"High-risk patients: {len(high_risk(patient_list))}")
    print(f"Total revenue:     EUR {sum(e['cost_eur'] for e in encounters_list):,.2f}")
    print()
    print("Top 3 most expensive encounters:")
    top_3 = sorted(encounters_list, key=lambda e: e["cost_eur"], reverse=True)[:3]
    for e in top_3:
        print(f"  {e['encounter_id']}  {e['ward']:>20s}  EUR {e['cost_eur']:>9,.2f}")
    print("=" * 50)

report(patients, encounters)
```

**Why this works:** This is the kind of code you might actually write in your project. Composition of small functions into a meaningful narrative output.
