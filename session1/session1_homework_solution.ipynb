## Task 1 — Three patients

Given cell, no changes. Just run it.

````python
patient_1 = {"id": "P101", "name": "Mila Okoye",     "age": 42, "weight_kg": 68.0, "height_m": 1.65}
patient_2 = {"id": "P102", "name": "Jonas Krüger",   "age": 59, "weight_kg": 94.5, "height_m": 1.78}
patient_3 = {"id": "P103", "name": "Beatrice Rossi", "age": 33, "weight_kg": 55.2, "height_m": 1.71}
````

## Task 2 — BMI for each patient

**Expected output format:**
```
Mila Okoye (ID P101): BMI = 25.0 kg/m^2
Jonas Krüger (ID P102): BMI = 29.8 kg/m^2
Beatrice Rossi (ID P103): BMI = 18.9 kg/m^2
```

Students are expected to use `**` for the square and `round(value, 1)` or an f-string format spec for one decimal place. Both approaches are acceptable.

````python
# Patient 1
bmi_1 = patient_1["weight_kg"] / (patient_1["height_m"] ** 2)
print(f'{patient_1["name"]} (ID {patient_1["id"]}): BMI = {bmi_1:.1f} kg/m^2')
````

````python
# Patient 2
bmi_2 = patient_2["weight_kg"] / (patient_2["height_m"] ** 2)
print(f'{patient_2["name"]} (ID {patient_2["id"]}): BMI = {bmi_2:.1f} kg/m^2')
````

````python
# Patient 3
bmi_3 = patient_3["weight_kg"] / (patient_3["height_m"] ** 2)
print(f'{patient_3["name"]} (ID {patient_3["id"]}): BMI = {bmi_3:.1f} kg/m^2')
````

### Alternative style: `round()` instead of format spec

Also correct — students who came from calculators often reach for this first:

````python
bmi_1 = round(patient_1["weight_kg"] / (patient_1["height_m"] ** 2), 1)
print(patient_1["name"] + " (ID " + patient_1["id"] + "): BMI = " + str(bmi_1) + " kg/m^2")
````

## Task 3 — Reflection (example answers)

These are *example* answers a student might give — not a grading rubric. The real goal is: did they notice the repetition? Did they think about scale? Are they honest about where they got stuck?

I kept worrying I would miss one of the `_2`s when I changed the numbers.

**2 · With 100 patients, what I would *not* want to do by hand?**

Definitely not copy-paste the cell 100 times. I also would not want to type 100 variable names. There must be a way to tell Python "do this thing once, but for every patient in a list."

**3 · What did I google / ask AI?**

- How to format a float to one decimal place in an f-string — found `{value:.1f}`.
- Whether `**` really means "to the power of" in Python, because in some languages it means XOR. Confirmed: `**` is exponent, `^` would be the XOR operator in Python.
- I asked an AI chatbot to "write the BMI line for me" once, to compare with what I had written myself.

## Task 4 (Optional) — Fast-Track: list + `for` loop

This is the payoff of the reflection above. It is also exactly what Session 2 opens with — no problem if students did not get here yet.

````python
patients = [patient_1, patient_2, patient_3]

for p in patients:
    bmi = p["weight_kg"] / (p["height_m"] ** 2)
    print(f'{p["name"]} (ID {p["id"]}): BMI = {bmi:.1f} kg/m^2')
````

## Teacher notes

### What to look for when grading

- **Correctness of the BMI values** (to one decimal place, with `**` as the exponent, not `^` — some students will try `^` because it is the math symbol).
- **Readable output format** — dictionary field names accessed correctly (students sometimes write `patient_1.name` from other languages).
- **Rounding** — one decimal place, no more. Students who print `25.041322314049586` have not yet learned format specs. Flag but do not penalise; this is exactly what Session 2 covers.
- **Reflection in Task 3** — any honest answer passes. What you are really checking: did they notice the repetition? If yes, the motivation for Session 2 is already there.

### Expected confusions (for your feedback loop)

| Symptom | What is going on | Session 2 fix |
|---|---|---|
| `^` used instead of `**` | mixing math notation with Python | demonstrate `2**3 == 8`, `2^3 == 1` (XOR) |
| Too many decimal places in output | not yet familiar with format specs | teach `{value:.1f}` vs `round(value, 1)` |
| Looking up a dict with `patient_1.name` | coming from other languages | remind them Python dicts use `["key"]`, not `.key` |
| Forgot to run the cell that defines the dicts | kernel state confusion | ties back to slide 21 Phase 5 DD |
