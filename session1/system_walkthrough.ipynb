## Cell 1 — Import a library

The line below says: *"Find the pandas library, load it into this kernel, and give me a short name for it: `pd`."*

This is **Path 3** from the slides in action.  
`pandas` is not part of Python itself — it was installed once, via `pip`, and now lives in your Python environment. This line just *activates* it inside this notebook's kernel.

```python
import pandas as pd
import matplotlib.pyplot as plt

print("pandas version:", pd.__version__)
```

**What happened?**

The kernel walked to your Python environment, found `pandas` and `matplotlib`, loaded them into its own memory (RAM), and printed which version of pandas it found.

Nothing on your *disk* has changed. Only this kernel's working memory now knows about pandas.

> If the kernel dies, this knowledge is gone. You'd have to `import` again when a new kernel starts.

---

## Cell 2 — Read data from the disk into the kernel

Now we take a CSV file that lives on your laptop — in `session1/data/sample_patients.csv` — and load its content into a **DataFrame**.

A DataFrame is just a table inside the kernel's memory. Think: *Excel sheet that only lives in RAM*.

**Path traced:**  
Disk (CSV file) → the kernel reads it → holds it in RAM under the name `df`

```python
df = pd.read_csv("data/sample_patients.csv")

print("Rows and columns:", df.shape)
```

**What happened?**

- The CSV is **still on your disk** — completely unchanged.
- A *copy* of it is now held in RAM by this kernel.
- That copy has a name: `df`. In the course we will read `df` as "the DataFrame".
- `df.shape` tells us: **30 rows, 8 columns**.

> Where does `df` live? **Only in the kernel's RAM.** Close this kernel and `df` disappears — the CSV on disk stays.

---

## Cell 3 — Look at the first five rows

We tell the kernel: *"Give me the top of `df` so I can see what this table actually contains."*

```python
df.head()
```

**What you see:**

30 synthetic patients (fake — no real data, by design). For each one: an ID, age, sex, height in cm, weight in kg, systolic and diastolic blood pressure, and whether they smoke (0 = no, 1 = yes).

> **Note:** the output above is not the DataFrame itself — it's a *rendering* of the first five rows that the kernel produced and the cell displayed. The DataFrame is in RAM; the table you see is a picture of it.

---

## Cell 4 — Ask for a quick summary

One line, and the kernel computes descriptive statistics for every numeric column: count, mean, min, max, quartiles.

```python
df.describe()
```

**What happened?**

The kernel looped through 30 patients, aggregated each numeric column, and rendered the result as a small summary table.

If you wanted to do this by hand in Excel, you'd set up formulas column by column. Here: **one line of code** tells a trained engine to do it all.

This is the power of a good library — and later in the course you will write your own tiny versions of this.

---

## Cell 5 — A first plot

Let's visualise the relationship between age and systolic blood pressure — a classic clinical plot.

```python
fig, ax = plt.subplots(figsize=(7, 4))

ax.scatter(df["age"], df["systolic_bp"], alpha=0.7)
ax.set_xlabel("Age (years)")
ax.set_ylabel("Systolic blood pressure (mmHg)")
ax.set_title("Systolic BP vs. age — 30 synthetic patients")
ax.grid(alpha=0.3)

plt.show()
```

**What happened?**

- The kernel pulled two columns out of `df` — `age` and `systolic_bp`.
- `matplotlib` drew a scatter plot in the kernel's memory.
- The image was sent back to the notebook and rendered below the cell.

If you had 30,000 patients instead of 30, this same line would still work — the *logic* doesn't change with size. That is why we are doing things this way.

> Clinical note: with only 30 synthetic rows the trend is noisy. With real epidemiological data you'd see the well-known age-BP gradient very clearly.

---

## Cell 6 — A little calculation

Add a new column to the DataFrame: BMI (body mass index). Formula: weight in kg divided by height in metres, squared.

You are *not* asked to understand the syntax. Just note: in a single line we create a new column that is computed from two existing ones, for **all 30 patients at once**.

```python
df["bmi"] = df["weight_kg"] / (df["height_cm"] / 100) ** 2

df[["patient_id", "age", "height_cm", "weight_kg", "bmi"]].head()
```

**What happened?**

- A new column `bmi` was added to the `df` that lives in the kernel's RAM.
- The **CSV on disk is still untouched**. If you re-open `data/sample_patients.csv` in a text editor, there is no `bmi` column there.
- To persist the new column, you would have to *explicitly* ask the kernel to write it back to disk (we'll learn that later in the course).

> This is the `.ipynb`-vs-RAM-vs-disk distinction from the slides, live. Three different places where something can live.
