### Variables — labels, not boxes

A variable name is a *label* attached to a value. The value lives in memory; the name is just a way to point at it.

````python
patient_name = "Anna Schmidt"
age_years = 54
height_m = 1.68
weight_kg = 79.5

print(patient_name, age_years, height_m, weight_kg)
````

````python
# Every value has a type. The label itself has no type — just the value behind it.
print(type(patient_name), type(age_years), type(height_m))
````

## Phase 4 — Clinic: BMI for a patient

First a single patient as a **dictionary** — a small container for key/value pairs. Today we just use it; we unpack dicts in Session 3.

````python
patient = {
    "id": "P001",
    "name": "Anna Schmidt",
    "age": 54,
    "weight_kg": 79.5,
    "height_m": 1.68,
}

patient
````

````python
# BMI = weight (kg) / height (m)^2
bmi = patient["weight_kg"] / (patient["height_m"] ** 2)
bmi
````

````python
# f-string for a clean, readable printout
print(f"{patient['name']} (ID {patient['id']}): BMI = {bmi:.1f} kg/m^2")
````

### Preview of Session 2 — a first decision

We won't dive in yet, but here is a taste of *control flow*. Don't worry if the syntax feels fresh — we spend all of Session 2 on it.

````python
if bmi < 18.5:
    category = "underweight"
elif bmi < 25:
    category = "normal weight"
elif bmi < 30:
    category = "overweight"
else:
    category = "obese"

print(f"{patient['name']}: BMI {bmi:.1f} → {category}")
````

## Differential diagnosis — typical Session-1 errors

- `NameError: name 'x' is not defined` → you are using a label that was never assigned (or you skipped a cell above).
- `SyntaxError` → usually a missing colon, missing bracket, or mixed quotes.
- `ZeroDivisionError` → dividing by zero. Check your input.
- `TypeError: unsupported operand type(s)` → you are mixing types, e.g. a number and a string.

If confused, add `print(type(x), x)` before the failing line. Nine times out of ten this reveals the cause.
