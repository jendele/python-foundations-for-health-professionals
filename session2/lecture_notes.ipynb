## Phase 2 — Anatomy: the building blocks

### Booleans and comparison

```python
True, False, type(True)
```

```python
# Comparison — note the double equals sign for equality
print(5 == 5)
print(5 != 4)
print(5 < 10)
print(5 <= 5)
print("a" == "A")   # case-sensitive
```

```python
# Logical operators: and, or, not
print(True and False)
print(True or False)
print(not True)
print((5 > 3) and (10 < 100))
```

### Conditional branching: `if` / `elif` / `else`

Python reads `if`/`elif`/`else` top to bottom. The first matching branch runs; the rest is skipped. **Order matters** — put the most specific (narrow) conditions first.

```python
bp_systolic = 190
patient_age = 12

if bp_systolic >= 160:
    print("Stage 2 hypertension")
    if patient_age < 25:
            print("Alert")
elif bp_systolic >= 140:
    print("Stage 1 hypertension")
elif bp_systolic >= 130:
    print("High-Normal")
else:
    print("Normal")
```

### `for` loops — iterate over a collection

```python
bp_readings = [118, 135, 142, 159, 168]

for bp in bp_readings:
    if bp >= 140:
        print(f"{bp}: high")
    else:
        print(f"{bp}: ok")
```

```python
# range() — note: end is EXCLUSIVE
for i in range(5):
    print(i)
```

### `while` loops — iterate on a condition

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

### `break` and `continue`

```python
# break — leave the loop immediately
for bp in bp_readings:
    if bp >= 160:
        print(f"Stop at {bp}!")
        break
    print(bp)
```

```python
# continue — skip the rest of THIS iteration, keep going
for bp in bp_readings:
    if bp < 140:
        continue
    print(f"High: {bp}")
```

## Phase 3 — Physiology: what actually happens

### Truthiness

In a boolean context, Python converts values automatically. **Empty = false. Zero = false. Everything else = true.**

```python
for value in [0, 1, -5, "", "hello", [], [0], None]:
    print(repr(value), "→", bool(value))
```

```python
# Idiomatic Python: check emptiness directly
some_list = [1, 2, 3]
if some_list:
    print("List has content")
```

### Short-circuit evaluation

`A and B`: if `A` is falsy, `B` is never evaluated.  
`A or B`: if `A` is truthy, `B` is never evaluated.  
This protects us from errors on the second half when the first half tells us we should stop.

```python
patient = None

# This does NOT error, because short-circuit stops at patient = None
if patient and patient["age"] > 65:
    print("senior")
else:
    print("no patient or not senior")
```

### List comprehensions — a for-loop that builds a list, in one line

```python
# Classic
high_bp_v1 = []
for bp in bp_readings:
    if bp >= 140:
        high_bp_v1.append(bp)

# Comprehension — same result, one line
high_bp_v2 = [bp for bp in bp_readings if bp >= 140]

print(high_bp_v1)
print(high_bp_v2)
```

## Phase 4 — Clinic: Blood pressure classification

Simplified ESH/ESC categorisation:

- **Normal:** systolic < 130 AND diastolic < 85
- **High-Normal:** systolic 130–139 OR diastolic 85–89
- **Stage 1:** systolic 140–159 OR diastolic 90–99
- **Stage 2:** systolic ≥ 160 OR diastolic ≥ 100

```python
patients = [
    {"id": "P001", "systolic": 118, "diastolic":  76},
    {"id": "P002", "systolic": 142, "diastolic":  92},
    {"id": "P003", "systolic": 168, "diastolic": 104},
    {"id": "P004", "systolic": 128, "diastolic":  82},
    {"id": "P005", "systolic": 155, "diastolic":  98},
    {"id": "P006", "systolic": 112, "diastolic":  70},
]
```

### Step 1 — classify one patient

```python
p = patients[0]
sys = p["systolic"]
dia = p["diastolic"]

if sys >= 160 or dia >= 100:
    category = "Stage 2"
elif sys >= 140 or dia >= 90:
    category = "Stage 1"
elif sys >= 130 or dia >= 85:
    category = "High-Normal"
else:
    category = "Normal"

print(f"{p['id']}: {sys}/{dia} → {category}")
```

### Step 2 — classify all patients in a loop

```python
for p in patients:
    sys = p["systolic"]
    dia = p["diastolic"]

    if sys >= 160 or dia >= 100:
        category = "Stage 2"
    elif sys >= 140 or dia >= 90:
        category = "Stage 1"
    elif sys >= 130 or dia >= 85:
        category = "High-Normal"
    else:
        category = "Normal"

    print(f"{p['id']}: {sys}/{dia} → {category}")
```

### Step 3 — count how many fall in each category

This is your first **aggregation** — reducing a list to a summary.

```python
counts = {"Normal": 0, "High-Normal": 0, "Stage 1": 0, "Stage 2": 0}

for p in patients:
    sys = p["systolic"]
    dia = p["diastolic"]

    if sys >= 160 or dia >= 100:
        category = "Stage 2"
    elif sys >= 140 or dia >= 90:
        category = "Stage 1"
    elif sys >= 130 or dia >= 85:
        category = "High-Normal"
    else:
        category = "Normal"

    counts[category] += 1

counts
```

## Differential diagnosis — common control-flow errors

- `=` vs `==`: `=` assigns, `==` compares. `if age = 18:` raises a SyntaxError in Python (nicer than C, which silently assigns).
- **Indentation errors**: the code inside `if`, `for`, `while` must be indented, always 4 spaces. Mixed tabs/spaces break this.
- **Off-by-one**: `range(5)` gives `0..4`, not `1..5`. For `1..5` use `range(1, 6)`.
- **Infinite while**: if the condition is never updated inside the loop, it runs forever. Kernel interrupt (stop button) is your friend.

Debugging shortcut: when a condition doesn't do what you expect, add `print(type(x), x)` before the `if`.
