# Mini-Challenge 1 — Patient Roster
### Saint Mary's · Monday morning, 09:00

> *"You are the receptionist on Monday morning. Five patients have been admitted overnight. Build a Python data structure that holds today's roster — and answer two questions about it."*

**Time:** 25 minutes · **Mode:** tandem · **Goal:** practise list-of-dicts, filtering, and a tiny aggregation.

---
## Brief

You receive these admissions on a paper sheet:

| Patient ID | Age | Gender | Ward             | LOS expected (days) |
|-----------|-----|-----|------------------|---------------------|
| P0101     | 78  | F   | Geriatrics       | 6                   |
| P0102     | 54  | M   | Cardiology       | 3                   |
| P0103     | 71  | F   | InternalMedicine | 5                   |
| P0104     | 39  | M   | Surgery          | 2                   |
| P0105     | 83  | F   | Geriatrics       | 9                   |

## Tasks
1. **Build** a list of 5 dictionaries called `roster` — one dict per patient, with the five fields above (use sensible key names: `patient_id`, `age`, `gender`, `ward`, `expected_los`).
2. **Filter:** create a new list `elderly` containing only patients aged 65 or older.
3. **Aggregate:** print the **count** of elderly patients and their **mean age**, formatted to one decimal.

> **Fast-Track (if you finish early):** Sort the original roster by `expected_los` *descending* — print the resulting list with one patient per line, formatted as `P0103 | LOS 5 | InternalMedicine`.

---

# Solution — Mini-Challenge 1

## Step 1 — Build the roster

```python
roster = [
    {"patient_id": "P0101", "age": 78, "gender": "F", "ward": "Geriatrics",       "expected_los": 6},
    {"patient_id": "P0102", "age": 54, "gender": "M", "ward": "Cardiology",       "expected_los": 3},
    {"patient_id": "P0103", "age": 71, "gender": "F", "ward": "InternalMedicine", "expected_los": 5},
    {"patient_id": "P0104", "age": 39, "gender": "M", "ward": "Surgery",          "expected_los": 2},
    {"patient_id": "P0105", "age": 83, "gender": "F", "ward": "Geriatrics",       "expected_los": 9},
]
print(len(roster), "patients in roster")
print(roster[0])
```

## Step 2 — Filter for patients aged 65+

```python
elderly = [p for p in roster if p["age"] >= 65]
print(len(elderly), "elderly patients")
```

## Step 3 — Count and mean age

```python
ages = [p["age"] for p in elderly]
mean_age = sum(ages) / len(ages)
print(f"Elderly count: {len(elderly)}")
print(f"Mean age: {mean_age}")
```

## Fast-Track

```python
sorted_roster = sorted(roster, key=lambda p: p["expected_los"], reverse=True)
for p in sorted_roster:
    print(f"{p['patient_id']} | LOS {p['expected_los']} | {p['ward']}")
```
