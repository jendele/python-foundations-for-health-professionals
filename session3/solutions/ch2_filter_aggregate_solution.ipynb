# Mini-Challenge 2 — The CFO needs a number, fast
### Saint Mary's · Tuesday afternoon, 15:30

> *"I'm in a board meeting in 20 minutes. How many cardiology cases did we have this quarter, and what did they cost us in total?"*

**Time:** 25 minutes · **Mode:** tandem · **Goal:** filter, count, and sum on a list of 50 encounters.

---

# Solution — Mini-Challenge 2

```python
# Test data — 50 encounters. Run this cell first.
import random
random.seed(7)
WARDS = ["Cardiology", "InternalMedicine", "Surgery", "ICU", "Geriatrics"]
DRGS  = ["F62B", "F71B", "G67B", "B70A", "I47B", "F49G", "K62B", "J64A"]
encounters = []
for i in range(50):
    encounters.append({
        "encounter_id": f"E{i+1:04d}",
        "ward": random.choice(WARDS),
        "drg_code": random.choice(DRGS),
        "los_days": random.randint(1, 14),
        "actual_cost_eur": round(random.uniform(900, 9500), 2),
    })
print(len(encounters), "encounters loaded")
print("Wards present:", set(e["ward"] for e in encounters))
```

## Step 1 — Cardiology only

```python
cardio = [e for e in encounters if e["ward"] == "Cardiology"]
print(len(cardio), "cardiology cases")
```

## Step 2 — The CFO's two numbers

```python
total_cost = sum(e["actual_cost_eur"] for e in cardio)
print(f"Cardiology cases: {len(cardio)}")
print(f"Total cost:       € {total_cost:,.2f}")
```

## Fast-Track — Total cost per DRG

```python
cost_by_drg = {}
for e in encounters:
    drg = e["drg_code"]
    cost_by_drg[drg] = cost_by_drg.get(drg, 0) + e["actual_cost_eur"]

# sorted from most expensive to least:
for drg, total in sorted(cost_by_drg.items(), key=lambda kv: kv[1], reverse=True):
    print(f"{drg}: € {total:,.2f}")
```
